```python
# Load Gemma 3n base model using Unsloth
from unsloth import FastModel
import torch

# Set sequence length and load mode
max_seq_length = 2048
load_in_4bit = True

# Use Unsloth's 4-bit Gemma 3n model checkpoint
model_checkpoint = "unsloth/gemma-3n-E2B-it-unsloth-bnb-4bit"

# Load the model and tokenizer
model, tokenizer = FastModel.from_pretrained(
    model_checkpoint,
    max_seq_length = max_seq_length,
    dtype = None,
    load_in_4bit = load_in_4bit,
    full_finetuning = False,
)

model.config.use_cache = False
model.config.pretraining_tp = 1
```

```python
# Load Preprocessed Audio Embedding Data
import pandas as pd

# Load training and validation data
train_df = pd.read_json("train.jsonl", lines=True)
val_df = pd.read_json("val.jsonl", lines=True)
print(train_df.head())
```

```python
# Dataset Wrapper for PyTorch
import torch
from torch.utils.data import Dataset

# Custom dataset class to wrap audio embeddings and labels
class AudioEmbeddingDataset(Dataset):
    def __init__(self, df):
        # Convert list of embeddings and label_ids into tensors
        self.embeddings = torch.tensor(df["embedding"].tolist(), dtype=torch.float32)
        self.labels = torch.tensor(df["label_id"].tolist(), dtype=torch.long)

    def __len__(self):
        return len(self.embeddings)

    def __getitem__(self, idx):
        return self.embeddings[idx], self.labels[idx]
```

```python
# Create dataset objects
train_dataset = AudioEmbeddingDataset(train_df)
val_dataset = AudioEmbeddingDataset(val_df)
```

```python
# DataLoader Setup
from torch.utils.data import DataLoader

batch_size = 16

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)
```

```python
# Define Classification Head
import torch.nn as nn

class GemmaAudioClassifier(nn.Module):
    def __init__(self, gemma_model, embedding_dim, num_classes):
        super().__init__()
        self.gemma = gemma_model
        self.classifier = nn.Sequential(
            nn.Linear(embedding_dim, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, num_classes)
        )

    def forward(self, audio_embedding):
        return self.classifier(audio_embedding)
```

```python
# Initialize classifier model
embedding_dim = len(train_df["embedding"].iloc[0])
num_classes = train_df["label_id"].nunique()

model = GemmaAudioClassifier(
    gemma_model=model,
    embedding_dim=embedding_dim,
    num_classes=num_classes
).to("cuda" if torch.cuda.is_available() else "cpu")

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4)
```

```python
from torch.utils.data import DataLoader
# Dataloaders for training and validation
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)
```

```python
import torch
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# One training epoch
def train_epoch(model, loader, loss_fn, optimizer):
    model.train()
    running_loss = 0.0
    correct, total = 0, 0

    for batch in tqdm(loader):
        x, y = batch
        x, y = x.to(device), y.to(device)

        optimizer.zero_grad()
        preds = model(x)
        loss = loss_fn(preds, y)
        loss.backward()
        optimizer.step()

        running_loss += loss.item() * x.size(0)
        predicted = torch.argmax(preds, dim=1)
        correct += (predicted == y).sum().item()
        total += y.size(0)

    return running_loss / total, correct / total
```

```python
# Evaluation step
def eval_model(model, loader, loss_fn):
    model.eval()
    running_loss = 0.0
    correct, total = 0, 0

    with torch.no_grad():
        for batch in loader:
            x, y = batch
            x, y = x.to(device), y.to(device)

            preds = model(x)
            loss = loss_fn(preds, y)

            running_loss += loss.item() * x.size(0)
            predicted = torch.argmax(preds, dim=1)
            correct += (predicted == y).sum().item()
            total += y.size(0)

    return running_loss / total, correct / total
```

```python
# Run training loop
EPOCHS = 10

for epoch in range(EPOCHS):
    train_loss, train_acc = train_epoch(model, train_loader, loss_fn, optimizer)
    val_loss, val_acc = eval_model(model, val_loader, loss_fn)

    print(f"\nEpoch {epoch+1}/{EPOCHS}")
    print(f"Train Loss: {train_loss:.4f}, Accuracy: {train_acc:.4f}")
    print(f"Val   Loss: {val_loss:.4f}, Accuracy: {val_acc:.4f}")
```

```python
import pandas as pd
import torch

# Load test data
test_df = pd.read_json("test.jsonl", lines=True)

# Convert embeddings to tensor
embeddings = torch.tensor(test_df["embedding"].tolist(), dtype=torch.float32)
embeddings = embeddings.to("cuda" if torch.cuda.is_available() else "cpu")

# Set model to eval
model.eval()
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# Predict
with torch.no_grad():
    outputs = model(embeddings)
    predicted_ids = torch.argmax(outputs, dim=1).cpu().numpy()

# Map predicted label IDs to emotions
label_map = {
    0: "Anger",
    1: "Disgust",
    2: "Fear",
    3: "Happiness",
    4: "Neutral",
    5: "Sadness"
}
test_df["predicted_label_id"] = predicted_ids
test_df["predicted_emotion"] = test_df["predicted_label_id"].map(label_map)

# Print results
print(test_df[["transcription", "predicted_emotion", "embedding"]].head())
```
